```python
def binary_search (arr,ele):
    #first and last index values
    first = 0
    last = len(arr) -1
    found = False
    while first <= last and not found:
        mid = int(first+last)/2
        # Match found 
        if arr[mid] == ele:
            found = True
        # Set new  midpoints up or down depending on comparision 
        else:
            # Set down
            if ele < arr[mid]:
                last = mid -1
            # Set up
            else:
                first = mid + 1
    return found 
```

```python
def rec_bin_search(arr,ele):
    #base Case!
    if len(arr) ==0:
        return False
    # Recursive Case
    else:
        mid = len(arr)/2
        #if match found
        if arr[mid]==ele:
            return True 
        else:
            # Call again on second half
            if ele<arr[mid]:
                return rec_bin_search(arr[:mid],ele)
            # Or call on first half
            else:
                return rec_bin_search(arr[mid+1:],ele)
```
